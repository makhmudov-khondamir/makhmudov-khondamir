# Title: **Downloading and Compressing the MNIST Sample Dataset**

This code snippet demonstrates how to download the MNIST sample dataset using Fastai in a Google Colab environment, compress it into a zip file, and download the compressed file to your local machine. It does not require a GPU and can be run in a CPU environment. This approach is useful for practicing with datasets in real-world scenarios where datasets are not pre-packaged in a library.

```python
from fastai.data.external import untar_data, URLs
import shutil
from google.colab import files

# Download the MNIST sample dataset
path = untar_data(URLs.MNIST_SAMPLE)
print(f"Dataset downloaded to: {path}")

# Compress the dataset directory into a zip file
shutil.make_archive('mnist_sample', 'zip', path)

# Download the zip file to your PC
files.download('mnist_sample.zip')
```

# Title: **Uploading and Extracting the MNIST Sample Dataset in Google Colab**

This code snippet guides you through the process of uploading the previously downloaded MNIST sample dataset (containing only digits 7 and 3) to Google Colab, extracting the contents from the zip file, and setting the path to the extracted files. This is useful for users who need to work with datasets in a controlled environment and prefer manual dataset management.

```python
import zipfile
import os

# Set the filename of the uploaded zip file
zip_file_path = "/content/mnist_sample.zip"

# Define the extraction path
extracted_folder_path = "/content/mnist_sample"

# Create the directory if it doesn't exist
os.makedirs(extracted_folder_path, exist_ok=True)

# Extract the zip file
with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    zip_ref.extractall(extracted_folder_path)
```
